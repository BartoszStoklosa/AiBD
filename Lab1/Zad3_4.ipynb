```python
import numpy as np
import matplotlib.pyplot as plt

#stworzenie wektorów z krokiem 0.1
x1 = np.arange(-1, 1, 0.1)
x2 = np.arange(-6, 6, 0.1)
x3 = np.arange(0, 5, 0.1)

# stworzenie funkncji
f = lambda x : x**2 + 5

#stworzenie wykresów
fig, axs = plt.subplots(1,3)
fig.set_size_inches(20, 6)
axs[0].plot(x1, f(x1))                 # wykres 1
axs[0].set_title('x^2 + 5, x [-1, 1]') # legenda    
axs[0].grid()                          # siatka
axs[0].set_xlabel('x')                 # tytuł osi x
axs[0].set_ylabel('y')                 # tytuł osi y
axs[1].plot(x2, f(x2))                 # -||-
axs[1].set_title('x^2 + 5, x [-6, 6]')
axs[1].grid()
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[2].plot(x3, f(x3))
axs[2].set_title('x^2 + 5, x [0, 5]')
axs[2].grid()
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
plt.show()
```
